# Optimizing Anti-Reflective Coatings for Solar Panels

Solar panels offer a renewable source of power which harnesses energy from the sun. The amount of power that can be generated by a modern solar panel is contigent on the amount of light that shines on its surface. Therefore, if the amount of light absorbed by the panel can be maximized, the power generation can be increased. However, when an electromagnetic wave is incident on a medium, there will be a component of the incident wave that is
transmitted into the medium and a component that is reflected. This behaviour is governed by
Fresnel’s equations:

<table>
$$
\begin{matrix} \Gamma_{TE} = \frac{\eta_2 cos \theta_i - \eta_1 cos \theta_t}{\eta_2 cos \theta_i + \eta_1 cos \theta_t} & 
\tau_{TE} = \frac{\eta_2 cos \theta_i}{\eta_2 cos \theta_i + \eta_1 cos \theta_t} \\ 
\Gamma_{TM} = \frac{\eta_2 cos \theta_t - \eta_1 cos \theta_i}{\eta_2 cos \theta_t + \eta_1 cos \theta_i} & 
\tau_{TM} = \frac{\eta_2 cos \theta_i}{\eta_2 cos \theta_t + \eta_1 cos \theta_i} 
\end{matrix}
$$
</table>

Where $\eta_1$ and $\eta_2$ are the refractive indices of the first and second media, and $\theta_i$ and $\theta_t$ are the incident and transmitted angels. A solar panel can only generate power with the light that is transmitted into its solar cell. As a result, any light that is reflected at the Air-Solar Panel interface will not be used to generate power. An anti-reflective coating can be applied to the surface of the panel to reduce the amount reflected light. This will ameliorate the power output of the panel  $$

### Assumptions:

To simplify the analysis, a few assunmptions are made. First, it will be assumed that all incoming electromagnetic waves are incident perpendicular to the surface of the panel. This assumption simplifies Fresnel's equations to:

<table>
$$
\begin{matrix} \Gamma = \frac{\eta_2 - \eta_1 }{\eta_2 + \eta_1 } & 
\tau = 1 + \Gamma 
\end{matrix}
$$
</table>

Next, the amount of solar radiation absorbed by the panel and converted into energy will be idealized. It will be assumed that the wavelength distribution is given by an idealized solar spectral irradiance. The idealized “blackbody” irradiance is modeled as follows:

$$
I(\lambda) = \frac{6.16 \times 10^{15}}{\lambda^5 (e^{\frac{2484}{\lambda}}-1)}
$$

Finally, the power generated by the cell will be computed by taking the product of the "blackbody" irradiance and the transmissivity for a give wavelength:

$$
P =  \int_{\lambda_1}^{\lambda_2} T(\lambda)I(\lambda) \,d\lambda 
$$

## Transfer Matrix Method

Single layer and multi-layer anti-reflectinve coating can be used to minimize reflection. However, traditional interface analysis becomes very difficult as the number of layers increase. Due to the increasing difficulty to calculate the reflectance and transmittance using only the Fresnel’s equations as the number of layers increases, the Transfer Matrix Method (TMM) was devised. This method decomposes the calculation into smaller components. The TMM can handle an arbitrarily sized multilayer system by describing the resulting transfer matrix as the product of the wave interactions at each respective layer as shown:

![title](fig/TMM.png)

This method allows for the analysis of the propagation of light through various media and
through an arbitrary number of layers. The electric field components denoted with a “+” in Figure 1.0 are the forward propagating component of the wave while the field components denoted with a “-” are the backward propagating components. The subscript assigned to each component denotes the medium that it is in. The prime symbol distinguishes the right-hand side of a medium from the left-hand side. While layers 1 to N are slabs of finite width, medium 0 and N+1 are assumed to be unbounded. In order to describe the transmission and reflection of the wave at each interface, a boundary condition is created. Using this detailed description of the multilayer system, the TMM can create a transfer matrix that can equate the unbounded region n0 to the other unbounded region N+1. In order to do so, the Transfer matrix will be built up using the interaction of light at each interface and the interaction of light in each medium. First, in order to describe the interaction of light at each interface, a dynamical matrix will be used as follows:

$$
\left(\begin{array}{cc} E^+_{m-1} \\E^-_{m-1} \end{array}\right) = Q_{m-1,m}
\left(\begin{array}{cc} E^+_{m} \\E^-_{m} \end{array}\right)
$$ 

The dynamical matrix $Q_{m-1,m}$ is a 2x2 matrix defined in terms of the reflection and transmission components of the $m^{th}$ layer.

$$
Q_{m-1,m} = \frac{1}{\tau_{m-1,m}} 
\begin{bmatrix}
1 & \Gamma_{m-1,m}\\
\Gamma_{m-1,m} & 1
\end{bmatrix}
$$

Since each layer has a finite size, the characteristic of how the wave propagates in each medium must also be described. The propagation matrix is used to describe this relationship for each layer.

$$
\left(\begin{array}{cc} E^{\prime +}_{m} \\E^{\prime -}_{m} \end{array}\right) = P_{m}
\left(\begin{array}{cc} E^+_{m} \\E^-_{m} \end{array}\right)
$$ 

Where,

<table>
$$
\begin{matrix} 
P_m = \begin{bmatrix}e^{j \delta_{m}} & 0\\0 & e^{j \delta_{m}}\end{bmatrix},
& 
\delta_m = \frac{2 \pi}{\lambda}n_m d_m 
\end{matrix}
$$
</table>

$\delta_m$ is the phase thickness of the $m^{th}$ layer. The physical thickness is $d_m$, $\lambda$ is the free-space wavelength, and $n_m$ is the refractive index of the medium. Using the dynamical matrices for each interface and the propagation matrix for each medium, the transfer matrix can be built by taking the product of these components.

$$
\left(\begin{array}{cc} E^+_{0} \\E^-_{0} \end{array}\right) = T
\left(\begin{array}{cc} E^+_{N+1} \\E^-_{N+1} \end{array}\right)
$$ 

The transfer matrix is:

$$
T = 
\begin{bmatrix}T_{1,1} & T_{1,2}\\T_{2,1} & T_{2,2}\end{bmatrix} =
Q_{0,1} \prod_{m=1}^{N}P_m Q_{m,m+1}
$$

Using the description of the multilayer system provided by the TMM, the reflection coefficient and the transmission coefficient can be written as follows:

$$
\Gamma = \frac{E_r}{E_i} = \frac{T_{2,1}}{T_{1,1}} 
$$


$$
\tau = \frac{E_t}{E_i} = \frac{1}{T_{1,1}}
$$

Assuming $E^-_0 = E^{\prime-}_{N+1}$, conservation of energy yields,

$$
\lvert S^+_0 \rvert = \lvert S^-_0 \rvert + \lvert S^+_{N+1} \rvert \Longrightarrow 
\frac{\lvert E^+_0 \rvert^2}{2\eta^*_0} = \frac{\lvert E^-_0 \rvert^2}{2\eta^*_0} + \frac{\lvert E^+_{N+1} \rvert^2}{2\eta^*_{N+1}}
$$

Then,

$$
\frac{\lvert E^+_0 \rvert^2}{2\eta^*_0} = \frac{\lvert \Gamma \rvert ^2 \lvert E^+_0 \rvert^2}{2\eta^*_0} + \frac{\lvert \tau \rvert ^2 \lvert E^+_{N+1} \rvert^2}{2\eta^*_{N+1}} \Longrightarrow 
\lvert \Gamma \rvert ^2 + \lvert \tau \rvert ^2 \frac{2\eta^*_0}{\eta^*_{N+1}} = 1
$$

For a lossless, nonmagnetic medium:

$$
\lvert \Gamma \rvert ^2 + \lvert \tau \rvert ^2 \frac{n_{N+1}}{n_0} = 1
$$

Therefore, reflectance and transmittance are defined as:

<table>
$$
\begin{matrix} R = \lvert \Gamma \rvert ^2 , & 
T = \lvert \tau \rvert ^2 \frac{n_{N+1}}{n_0} 
\end{matrix}
$$
</table>
